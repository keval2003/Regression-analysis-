```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import cm
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('year_wise_indian_crop_production.csv')

color = cm.rainbow(np.linspace(0, 1, 3))
for i,c in zip(range(3),color):
  X = np.array([[1950],[1960],[1970],[1980],[1990],[2000],[2010],[2011],[2012],[2013],[2014],[2015],[2016],[2017]])
  Y = np.array(dataset.iloc[i:i+1,1:15].values)
  Y = Y.reshape((-1, 1))
  poly_reg = PolynomialFeatures(degree = 3)
  X_poly = poly_reg.fit_transform(X)
  lin_reg = LinearRegression()
  lin_reg.fit(X_poly, Y)
  plt.scatter(X, Y, color = c)
  # c=next(color)
  lab=dataset.iloc[i,0]
  plt.plot(X, lin_reg.predict(poly_reg.fit_transform(X)), color = c, label=dataset.iloc[i,0])
  plt.legend()
  plt.title('Crops production')
  plt.xlabel('Year: 1950->2017')
  plt.ylabel('Production (in half million tons)')
  # plt.show()
  print(lin_reg.predict(poly_reg.fit_transform([[2019]])))
```
